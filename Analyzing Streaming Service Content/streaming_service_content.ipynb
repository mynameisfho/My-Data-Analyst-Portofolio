## Preparing the data
### Merging Tables
Since most tables share the same column names, consolidate the data by using a series of [`UNION`](https://www.postgresql.org/docs/14/queries-union.html) statements. 

Use `UNION ALL` specifically to retain any duplicate rows, as content appearing on multiple platforms should be counted multiple times.

```python
-- 6
SELECT *
FROM amazon
UNION ALL
SELECT *
FROM hulu
UNION ALL
SELECT * 
FROM netflix
UNION ALL
SELECT *
FROM disney
```

One drawback of this method is that we lose information about the streaming service source. To address this, revise the query to include the streaming service information.

```python
-- 7
SELECT *, 'amazon' AS service
FROM amazon
UNION
SELECT *, 'hulu' AS service
FROM hulu
UNION
SELECT *, 'netflix' AS service
FROM netflix
UNION
SELECT *, 'disney' AS service
FROM disney
```

**Incorporating Genre Information**

The genres table could provide valuable context, so let’s add it by using a `JOIN`. This will require the use of a [Common Table Expression](https://www.postgresql.org/docs/current/queries-with.html), or CTE.

```python
-- 8
WITH service_data AS (
	SELECT *, 'amazon' AS service
    FROM amazon
    UNION
    SELECT *, 'hulu' AS service
    FROM hulu
    UNION
    SELECT *, 'netflix' AS service
    FROM netflix
    UNION
    SELECT *, 'disney' AS service
    FROM disney
)

SELECT *
FROM service_data AS sd
LEFT JOIN genres AS g
	ON sd.title = g.film
```

### Identifying Missing Data
We’ve noticed some missing values in the `age` and `imdb` columns and should also check the `rotten_tomatoes` column, as we may use it later. To quantify the missing values in each column, we can use a combination of `SUM()` and `CASE WHEN` statements.

To calculate the null values per column, use a combination of [SUM()](https://www.postgresql.org/docs/8.2/functions-aggregate.html) and [CASE WHEN](postgresql.org/docs/current/functions-conditional.html) to count the number of null values.

```python
-- 9
WITH service_data AS (
	SELECT *, 'amazon' AS service
    FROM amazon
    UNION
    SELECT *, 'hulu' AS service
    FROM hulu
    UNION
    SELECT *, 'netflix' AS service
    FROM netflix
    UNION
    SELECT *, 'disney' AS service
    FROM disney
),

all_data AS (
    SELECT *
    FROM service_data AS sd
    LEFT JOIN genres AS g
        ON sd.title = g.film
)

SELECT 
	SUM(CASE WHEN imdb IS NULL THEN 1 ELSE 0 END) AS imdb_missing,
    SUM(CASE WHEN age IS NULL THEN 1 ELSE 0 END) AS age_missing,
    SUM(CASE WHEN rotten_tomatoes IS NULL THEN 1 ELSE 0 END) AS rt_missing
FROM all_data
```

## Data Analysis
### Most Family-Friendly Streaming Service
To determine which streaming service is most family-friendly, we’ll calculate the percentage of children’s content on each platform. While we could rely on the primary genre column, a more robust approach is to use pattern matching to identify mentions of “kids,” “family,” and similar terms.

We have our primary genre column, but that could leave out some content. A better way may be to use [pattern matching](https://www.postgresql.org/docs/current/functions-matching.html) to find any references to "kids", "family", etc.

```python
-- 10
WITH service_data AS (
	SELECT *, 'amazon' AS service
    FROM amazon
    UNION
    SELECT *, 'hulu' AS service
    FROM hulu
    UNION
    SELECT *, 'netflix' AS service
    FROM netflix
    UNION
    SELECT *, 'disney' AS service
    FROM disney
),

all_data AS (
    SELECT *
    FROM service_data AS sd
    LEFT JOIN genres AS g
        ON sd.title = g.film
)

SELECT
	*
FROM all_data
WHERE genre ILIKE '%kids%' 
    OR genre ILIKE '%family%' 
    OR genre ILIKE '%children%'
```

Once confirmed, we’ll adjust our query, using `CASE WHEN` to perform aggregation and identify the platform with the highest percentage of family-friendly content.

```python
-- 11
WITH service_data AS (
	SELECT *, 'amazon' AS service
    FROM amazon
    UNION
    SELECT *, 'hulu' AS service
    FROM hulu
    UNION
    SELECT *, 'netflix' AS service
    FROM netflix
    UNION
    SELECT *, 'disney' AS service
    FROM disney
),

all_data AS (
    SELECT *
    FROM service_data AS sd
    LEFT JOIN genres AS g
        ON sd.title = g.film
)

SELECT
	service,
    AVG(CASE WHEN genre ILIKE '%kids%' 
        OR genre ILIKE '%family%' 
        OR genre ILIKE '%children%' THEN 1.0000 ELSE 0.0000 END) * 100 AS pct_family
FROM all_data
GROUP BY service
ORDER BY pct_family DESC
```

### Highest-Rated Content
Using the `rotten_tomatoes` column, we can examine content ratings by extracting the numeric value with [`SPLIT_PART()`](https://www.postgresql.org/docs/9.1/functions-string.html) and casting it as numeric data (`::`). We'll further segment the data by movies and TV shows and visualize this in a grouped bar chart.

```python
-- 12
WITH service_data AS (
	SELECT *, 'amazon' AS service
    FROM amazon
    UNION
    SELECT *, 'hulu' AS service
    FROM hulu
    UNION
    SELECT *, 'netflix' AS service
    FROM netflix
    UNION
    SELECT *, 'disney' AS service
    FROM disney
),

all_data AS (
    SELECT *
    FROM service_data AS sd
    LEFT JOIN genres AS g
        ON sd.title = g.film
)

SELECT
	service,
    CASE WHEN type = 1 THEN 'TV' ELSE 'Movie' END AS type,
    AVG(SPLIT_PART(rotten_tomatoes, '/', 1)::NUMERIC) AS rt_score
FROM all_data
GROUP BY service, type
ORDER BY rt_score DESC
```

### Comparing Critics and Audience Ratings Over Time
For our final analysis, let’s examine trends in critics’ and audience ratings of TV shows over time, setting aside the streaming service. 

To prepare the data for visualization, we’ll use [`TO_DATE()`](https://www.postgresql.org/docs/current/functions-formatting.html) to convert the release year to a date format.

```python
-- 13
WITH service_data AS (
	SELECT *, 'amazon' AS service
    FROM amazon
    UNION
    SELECT *, 'hulu' AS service
    FROM hulu
    UNION
    SELECT *, 'netflix' AS service
    FROM netflix
    UNION
    SELECT *, 'disney' AS service
    FROM disney
),

all_data AS (
    SELECT *
    FROM service_data AS sd
    LEFT JOIN genres AS g
        ON sd.title = g.film
)

SELECT
	date,
    AVG(ABS(imdb_score - rt_score)) AS avg_difference
FROM (
    SELECT 
        TO_DATE(year::TEXT, 'YYYY') AS date,
        SPLIT_PART(rotten_tomatoes, '/', 1)::NUMERIC AS rt_score,
        SPLIT_PART(imdb, '/', 1)::NUMERIC * 10 AS imdb_score
    FROM all_data
    WHERE imdb IS NOT NULL
        AND rotten_tomatoes IS NOT NULL
    	AND year >= 2000
) AS sub
GROUP BY date
ORDER BY date
```
