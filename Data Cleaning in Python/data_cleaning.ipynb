# Data Cleaning in Python

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. Loading and Reviewing the Dataset
Work with a dataset of audiobooks downloaded from [audible dataset](https://www.kaggle.com/datasets/snehangsude/audible-dataset), covering releases from 1998 through pre-planned launches in 2025. Let's start by loading the raw data.

**Instructions:**
- Use pandas to load the `audible_raw.csv` file found in the data directory into a DataFrame named `audible`.
- Display the first few rows of `audible` to inspect the data.

```python
# Load the audible_raw.csv file
audible = pd.read_csv("data/audible_raw.csv")

# View the first rows of the dataframe
audible
```

##### The Data

- "name" - The name of the audiobook.
- "author" - The audiobook's author.
- "narrator" - The audiobook's narrator.
- "time" -  The audiobook's duration, in hours and minutes.
- "releasedate" -  The date the audiobook was published.
- "language" -  The audiobook's language.
- "stars" -  The average number of stars (out of 5) and the number of ratings (if available).
- "price" -  The audiobook's price in INR (Indian Rupee).

Use the `.info()` method to inspect the data types of the columns

```python
# Inspect the columns' data types
audible.info()
```

#### 2. Cleaning Text Data in Author and Narrator Columns
Clean up the author and narrator columns by removing "Writtenby:" and "Narratedby:" from the respective fields. Use the `.str.replace()` method.

**Instructions:**
- Remove 'Writtenby:' from the `author` column.
- Remove 'Narratedby:' from the `narrator` column.
- Display the changes to verify.

```python
# Remove Writtenby: from the author column
audible['author'] = audible['author'].str.replace('Writtenby:', '')

# Remove Narratedby: from the narrator column
audible['narrator'] = audible['narrator'].str.replace('Narratedby:', '')

# Check the results
audible
```

#### 3. Splitting the Stars Column into Ratings and Reviews Count
The `stars` column includes both the average rating and the count of reviews. Split this into two separate columns: `rating_stars` and `n_ratings`.

To start, examine some sample entries in the stars column using `.sample()`.

```python
# Get a glimpse of the stars column
audible.stars.sample(n=10)
```

Since there are instances labeled `Not rated yet`, filter these out before sampling again.


```python
# Explore the values of the star column that are not 'Not rated yet'
audible[audible.stars != 'Not rated yet'].stars.sample(n=10)
```

Next, replace `Not rated yet` with `NaN`.

```python
# Replace 'Not rated yet' with NaN
audible.stars.replace('Not rated yet', np.nan, inplace=True)
```

And use `.str.extract()` to get the number of stars and the number of ratings into their own columns.

**Instructions:**
- Extract the number of stars into `rating_stars`
- Extract the number of ratings into `n_ratings`
- Convert both columns to float data types.

```python
# Extract number of stars into rating_stars and turn into float
audible['rating_stars'] = audible['stars'].str.extract('^([\d.]+)').astype(float)

# Replace the comma, extract number of ratings into n_ratings and turn into float
audible['n_ratings'] = audible['stars'].str.replace(',', '').str.extract('(\d+) rating').astype(float)

# Examine the new rating_stars and n_ratings columns
audible[['rating_stars', 'n_ratings']]
```

Lastly, remove the original `stars` column with `.drop`.

```python
# Dorp the stars column
audible.drop('stars', axis=1, inplace=True)

# Check the results
audible
```

#### 4. Adjusting Data Types
It’s important to set columns to the correct data types for analysis.

**Instructions:**
- Convert `price` to a float.
- Set `rating_stars` as a categorical data type.
- Convert `releasedate` to a datetime format.

```python
# Explore the price column
audible.price.sample(n=10)
```

Before converting price, remove any commas or the word `Free` from the entries.

```python
# Replace the comma with ''
audible['price'] = audible.price.str.replace(',', '')

# Replace 'Free' with 0
audible['price'] = audible.price.str.replace('Free', '0')

# Turn price to float
audible['price'] = audible.price.astype(float)
```

Convert `rating_stars` to a categorical type.

Since `stars` has a limited set of discrete values, the best data type for the column is `category`.

First, examine the unique values in this column to confirm.



```python
# Look at the unique values in the rating_stars column
audible.rating_stars.unique()
```

Then, use `.astype` for the type change.


```python
# Turn rating_stars to category
audible['rating_stars'] = audible.rating_stars.astype('category')
```

To convert `releasedate`, apply the `to_datetime()` function to turn the dates into datetime objects

```python
# Convert releasedate to datetime
audible['releasedate'] = pd.to_datetime(audible.releasedate)

# Inspect the dataframe 
audible.info()
```

#### 5. Extracting Hours and Minutes from the `time` Column
The `time` column combines hours and minutes. Separate these and create a `time_minutes` column with total minutes.

```python
# Explore the values in the time column
audible.time.sample(n=10)
```

See the other ways to encoded `min` or `minutes`:

```python
# Search the entries in the time column for different spellings of min. Let' try min, mins, minutes
audible.time[audible.time.str.contains('min')].sample(n=10)
```

To normalize the text:
- Standardize "hr" and "hrs" to `hr`
- Standardize "min" and "mins" to `min`
- Convert "Less than 1 minute" to 1 min.

```python
# Replace hrs, mins, and 'Less than 1 minute'
audible['time'] = audible.time.str.replace('hrs', 'hr')
audible['time'] = audible.time.str.replace('mins', 'min')
audible['time'] = audible.time.str.replace('Less than 1 minute', '1 min')
```

Let's see how it looks now:

```python
# Check the results
audible[['time']]
```

The next step is to extract the hours and minutes from the text, then merge them into a new column `time_mins`.

**Instructions:** 
- Extract the number of hours from `time` into the `hours` variable.
- Extract the number of minutes from `time` into the `mins` variable.
- Create the `time_mins` by combining hours and minutes.

```python
# Extract the number of hours, turn to integer
hours = audible['time'].str.extract('(\d+) hr').fillna(0).astype(int)

# Extract the number of minutes, turn to integer
mins = audible['time'].str.extract('(\d+) min').fillna(0).astype(int)

# Combine hours and minutes into the time_mins column
audible['time_mins'] = 60 * hours + mins

# Check the results
audible[['time_mins']]
```

Lastly, delete the columns that we don't need any more

```python
# Drop the time column
audible.drop(columns=['time'], axis=1, inplace=True)
```

Here is the dataframe looks now:

```python
# Inspect the dataframe 
audible.info()
```

#### 7. Identifying and Removing Duplicates

Check for duplicates in the data.

Use `.duplicated()` combined with `.sum()` to count duplicates.

```python
# Look for duplicate rows
audible.duplicated().sum()
```

Then, check for duplicates in a specific subset of columns (`name`, `author`, `narrator`, `time_mins`, and `price`).

Use `.duplicated()` again, but with our subset of columns.

```python
# Create a list of our subset columns and assign to subset_cols
subset_cols=['name', 'author', 'narrator', 'time_mins', 'price']

# Check for duplicates using our subset of columns
audible.duplicated(subset=subset_cols).sum()
```

Look at those values (use `keep=false`) and see what is going on:

```python
# Check the duplicated rows keeping the duplicates and order by the name column
audible[audible.duplicated(subset=subset_cols, keep=False)].sort_values(by=['name'])
```

The duplicates are for files with different release dates. Keep the record with the last release date.

Use the `drop_duplicates()` method with the same subset and using `keep=last` to keep the last release date

```python
# Drop duplicated rows keeping the last release date
audible.drop_duplicates(subset=subset_cols, keep='last', inplace=True)
```

Check again for duplicates:

```python
# Check again for duplicates using our subset of columns
audible.duplicated(subset=subset_cols).sum()
```

#### 8. Handling Missing Values

Lastly, let’s identify any missing values.

Use `.isna()` with `.sum()` to count missing values by column

```python
# Check for null values
audible.isna().sum()
```

We could turn the **NaN** values to 0 or another numeric value, or we could keep them. It depends on our use case.

If we want to plot the ratings distribution, it can make sense to drop audiobooks with no ratings. But if we need to use the distribution of prices for our analysis, then removing audiobooks with no ratings will bias our results (since unrated audiobooks are likely more niche and might have a different pricing structure than rated audiobooks).

We will keep the unrated audiobooks for now.

Finally, we’ll save the cleaned dataset using `.to_csv` with `index=False` to avoid copying the current index into the output file.
